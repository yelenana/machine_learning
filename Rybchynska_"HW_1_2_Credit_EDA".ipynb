# Робота з пропущеними значеннями

Найпростіше емпіричне правило для опрацювання пропущених значень: якщо пропущених значень більше за 40% - видаляємо колонку, якщо менше за 40% - аналізуємо, як можна заповнити і чи треба.

Зазвичай, якщо відсоток пропущених даних більший за 10 і немає чіткої (яка значно виділяється) моди (найчастішого) значення в даних, то лишаємо дані як є до використання методів машинного навчання. Далі якщо метод вимагає заповення пропущених значень, можемо експериментувати із заповеннями. Також хорошою практикою є створити окрему колонку-флаг (0/1), яка вказує, де були пропущені значення - це буде додаткова ознака для моделі.

# **Виконання ДЗ2**

Функція для аналізу стовпця:

```python
def analyze_column_full(column_name, df, null_df):
    """
    Виводить детальну інформацію для заданого стовпця,
    включаючи пропущені значення, розподіл, боксплот,
    квантилі, моду, середнє та медіану.

    Args:
      column_name (str): Назва стовпця для аналізу.
      df (pd.DataFrame): Основний DataFrame з даними.
      null_df (pd.DataFrame): DataFrame з інформацією про пропущені значення.
    """

    print(f"*** Аналіз стовпця: {column_name} ***")

    # Вивід інформації про пропущені значення
    null_info = null_df[null_df['column_name'] == column_name]
    print(f"\nІнформація про пропущені значення для {column_name}:")
    print(null_info)
    if not null_info.empty:
      print(f"- Пропущені значення = {null_info['null_percentage'].iloc[0]:.2f}%")


    # Вивід розподілу значень
    print(f"\nРозподіл значень для {column_name}:")
    print(df[column_name].value_counts().head())

    # Побудова боксплоту
    print(f"\nБоксплот для {column_name}:")
    plt.figure(figsize = (6,4))
    sns.boxplot(df[column_name])
    plt.title(f'Боксплот для {column_name}')
    plt.show()

    # Розрахунок та вивід квантилів
    print(f"\nПроцентильні значення для {column_name}:")
    quantiles = df[column_name].quantile(q = [0.25,0.5,0.75,1])
    print(quantiles)

    # Розрахунок та вивід моди
    print(f"\nНайбільш повторюване значення в {column_name}:")
    mode_value = df[column_name].mode()
    if not mode_value.empty:
        print(mode_value[0])
    else:
        print("Мода не знайдена (можливо, всі значення унікальні)")

    # Розрахунок та вивід середнього
    print(f"\nСереднє значення {column_name}:")
    print(df[column_name].mean())

    # Розрахунок та вивід медіани
    print(f"\nМедіана стовпчику {column_name}:")
    print(df[column_name].median())

    print(f"\n*** Кінець аналізу для стовпця: {column_name} ***")
    print("-" * 30)
```

### Створення функції `age_cat` для категоризації `YEARS_BORNING`

```python
def age_cat(years):
    if years <= 20:
        return '0-20'
    elif years > 20 and years <= 30:
        return '20-30'
    elif years > 30 and years <= 40:
        return '30-40'
    elif years > 40 and years <= 50:
        return '40-50'
    elif years > 50 and years <= 60:
        return '50-60'
    elif years > 60 and years <= 70:
        return '60-70'
    elif years > 70:
        return '70+'
```

### Одномірний (категоріальний vs цільовий) та двомірний (категоріальний vs категоріальний) аналіз (гістограма) - категоріальні стовпчики

```python
import matplotlib.pyplot as plt


def bi_cat_countplot(df, column, hue_column):
    unique_hue_values = df[hue_column].unique()
    fig, axes = plt.subplots(nrows=1, ncols=2)
    fig.set_size_inches(14,6)

    pltname = f'Нормалізований розподіл значень за категорією: {column}'
    proportions = df.groupby(hue_column)[column].value_counts(normalize=True)
    proportions = (proportions*100).round(2)
    ax = proportions.unstack(hue_column).sort_values(
        by=unique_hue_values[0], ascending=False
        ).plot.bar(ax=axes[0], title=pltname)

    # анотація значень в барплоті
    for container in ax.containers:
        ax.bar_label(container, fmt='{:,.1f}%')

    pltname = f'Кількість даних за категорією: {column}'
    counts = df.groupby(hue_column)[column].value_counts()
    ax = counts.unstack(hue_column).sort_values(
        by=unique_hue_values[0], ascending=False
        ).plot.bar(ax=axes[1], title=pltname)

    for container in ax.containers:
      ax.bar_label(container)


def uni_cat_target_compare(df, column):
    bi_cat_countplot(df, column, hue_column='TARGET' )


def bi_countplot_target(df0, df1, column, hue_column):
  pltname = 'Клієнт зі складнощами щодо платності'
  print(pltname.upper())
  bi_cat_countplot(df1, column, hue_column)
  plt.show()

  pltname = 'Клієнти зі своєчасними платежами'
  print(pltname.upper())
  bi_cat_countplot(df0, column, hue_column)
  plt.show()
```

### Обчислення мінімального та максимального діапазону відхилень для числових стовпців

```python
def outlier_range(dataset,column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    Min_value = (Q1 - 1.5 * IQR)
    Max_value = (Q3 + 1.5 * IQR)
    return Max_value
```

# Робота з викидами для числових стовпців

```python
import warnings

def dist_box(dataset, column):
    with warnings.catch_warnings():
      warnings.simplefilter("ignore")

      plt.figure(figsize=(16,6))

      plt.subplot(1,2,1)
      sns.distplot(dataset[column], color = 'purple')
      pltname = 'Графік розподілу для ' + column
      plt.ticklabel_format(style='plain', axis='x')
      plt.title(pltname)

      plt.subplot(1,2,2)
      red_diamond = dict(markerfacecolor='r', marker='D')
      sns.boxplot(y = column, data = dataset, flierprops = red_diamond)
      pltname = 'Боксплот для ' + column
      plt.title(pltname)

      plt.show()
```

```python
def analyze_column_outliers(df0: pd.DataFrame, df1: pd.DataFrame, column: str):
    """
    Виконує аналіз викидів для заданого числового стовпця, розділеного за TARGET.
    Обчислює межі викидів за методом IQR для обох груп (0 та 1)
    та будує KDE графік розподілу без викидів.

    Args:
        df0 (pd.DataFrame): DataFrame з даними для TARGET == 0 (вчасні оплати).
        df1 (pd.DataFrame): DataFrame з даними для TARGET == 1 (платіжні труднощі).
        column (str): Назва числового стовпця для аналізу.
    """

    print(f"### Аналіз стовпця: `{column}`")

    # Перевірка наявності стовпця в обох датафреймах
    if column not in df0.columns or column not in df1.columns:
        print(f"Помилка: Стовпець '{column}' не знайдено в одному або обох датафреймах (df0, df1).")
        return

    # Перевірка, чи є стовпець числовим
    if not pd.api.types.is_numeric_dtype(df0[column]) or not pd.api.types.is_numeric_dtype(df1[column]):
         print(f"Помилка: Стовпець '{column}' не є числовим.")
         return

    #### Пошук викидів при випадках із платіжними труднощами (TARGET == 1)
    print(f"\n#### Пошук викидів в `{column}` при випадках із платіжними труднощами")
    Q1_1 = df1[column].quantile(0.25)
    Q3_1 = df1[column].quantile(0.75)
    IQR_1 = Q3_1 - Q1_1
    print(f"IQR для випадків із платіжними труднощами: {IQR_1:.2f}")

    Min_value1 = (Q1_1 - 1.5 * IQR_1)
    Max_value1 = (Q3_1 + 1.5 * IQR_1)
    print(f"Мінімальне значення, до якого існують викиди (платіжні труднощі): {Min_value1:.2f}")
    print(f"Максимальне значення, після якого існують викиди (платіжні труднощі): {Max_value1:.2f}")

    #### Пошук викидів при випадках із вчасними оплатами (TARGET == 0)
    print(f"\n#### Пошук викидів в `{column}` при випадках із вчасними оплатами")
    Q1_0 = df0[column].quantile(0.25)
    Q3_0 = df0[column].quantile(0.75)
    IQR_0 = Q3_0 - Q1_0
    print(f"IQR для випадків із вчасними оплатами: {IQR_0:.2f}")

    Min_value0 = (Q1_0 - 1.5 * IQR_0)
    Max_value0 = (Q3_0 + 1.5 * IQR_0)
    print(f"Мінімальне значення, до якого існують викиди (вчасні оплати): {Min_value0:.2f}")
    print(f"Максимальне значення, після якого існують викиди (вчасні оплати): {Max_value0:.2f}")

    #### Видалення викидів і побудова діаграми розподілу
    print("\n#### Побудова діаграми розподілу без викидів")

    plt.figure(figsize = (14,6))

    # Фільтруємо дані для побудови графіку, залишаючи тільки значення до Max_value
    # Викиди знизу (менше Min_value) зазвичай менш критичні для KDE, але їх також можна відфільтрувати,
    # хоча в даному випадку зосереджуємося на верхніх викидах, як у Вашому прикладі.
    # Якщо потрібно відфільтрувати обидва типи викидів, можна використовувати:
    # df1_filtered = df1[(df1[column] >= Min_value1) & (df1[column] <= Max_value1)]
    # df0_filtered = df0[(df0[column] >= Min_value0) & (df0[column] <= Max_value0)]

    df1_filtered = df1[df1[column] <= Max_value1].copy()
    df0_filtered = df0[df0[column] <= Max_value0].copy()

    if not df1_filtered.empty:
        sns.kdeplot(df1_filtered[column], label = 'Платіжні труднощі')
    else:
        print(f"Відсутні дані для KDEplot (платіжні труднощі) після фільтрації викидів для стовпця '{column}'.")

    if not df0_filtered.empty:
        sns.kdeplot(df0_filtered[column], label = 'Вчасні оплати')
    else:
         print(f"Відсутні дані для KDEplot (вчасні оплати) після фільтрації викидів для стовпця '{column}'.")


    plt.ticklabel_format(style='plain', axis='x')
    plt.xticks(rotation = 45)
    plt.title(f'Розподіл {column} без викидів')
    plt.xlabel(column)
    plt.ylabel('Щільність')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"*** Аналіз стовпця `{column}` завершено. ***")
    print("-" * 40)

# Приклад використання функції для стовпця 'AMT_CREDIT':
# Припускаючи, що df0 та df1 вже створені
# analyze_column_outliers(df0, df1, 'AMT_CREDIT')

# Приклад використання для іншого числового стовпця, наприклад 'AMT_INCOME_TOTAL':
# analyze_column_outliers(df0, df1, 'AMT_INCOME_TOTAL')

# Приклад використання для стовпця 'DAYS_BIRTH':
# analyze_column_outliers(df0, df1, 'DAYS_BIRTH')
```

Функція для виконання аналізу:

```python
def analyze_and_plot_columns(df1, df0, col1, col2, outlier_range_func):
    """
    Analyzes and plots the relationship between two numerical columns
    for two different groups (e.g., based on payment behavior), handling outliers.

    Args:
        df1 (pd.DataFrame): DataFrame for the first group (e.g., payment difficulties).
        df0 (pd.DataFrame): DataFrame for the second group (e.g., on-time payments).
        col1 (str): The name of the first column to analyze and plot.
        col2 (str): The name of the second column to analyze and plot.
        outlier_range_func (function): A function that takes a DataFrame and a column name
                                      and returns the maximum value to consider before outliers.
    """

    # Пошук викидів для стовпця 1 для групи із платіжними труднощами
    max_value1_col1 = outlier_range_func(df1, col1)
    print(f"Максимальне значення для {col1} у групі з платіжними труднощами (без викидів): {max_value1_col1}")

    # Пошук викидів для стовпця 2 для групи із платіжними труднощами
    max_value1_col2 = outlier_range_func(df1, col2)
    print(f"Максимальне значення для {col2} у групі з платіжними труднощами (без викидів): {max_value1_col2}")

    # Пошук викидів для стовпця 1 для групи із вчасними оплатами
    max_value0_col1 = outlier_range_func(df0, col1)
    print(f"Максимальне значення для {col1} у групі з вчасними оплатами (без викидів): {max_value0_col1}")

    # Пошук викидів для стовпця 2 для групи із вчасними оплатами
    max_value0_col2 = outlier_range_func(df0, col2)
    print(f"Максимальне значення для {col2} у групі з вчасними оплатами (без викидів): {max_value0_col2}")

    # Побудова діаграми розсіювання для порівняння з видаленими викидами
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.title('Payment difficulties')
    sns.scatterplot(x=df1[df1[col1] < max_value1_col1][col1],
                    y=df1[df1[col2] < max_value1_col2][col2], data=df1)
    plt.ticklabel_format(style='plain', axis='x')
    plt.ticklabel_format(style='plain', axis='y')
    plt.xlabel(col1)
    plt.ylabel(col2)

    plt.subplot(1, 2, 2)
    plt.title('On-Time Payments')
    sns.scatterplot(x=df0[df0[col1] < max_value0_col1][col1],
                    y=df0[df0[col2] < max_value0_col2][col2], data=df0)
    plt.ticklabel_format(style='plain', axis='x')
    plt.ticklabel_format(style='plain', axis='y')
    plt.xlabel(col1)
    plt.ylabel(col2)

    plt.tight_layout(pad=4)
    plt.show()
```

```python
def draw_boxplot(df, categorical, continuous, max_continuous, title, hue_column, subplot_position):
    """
    Малює блок-діаграму для заданого DataFrame, категоріальної та неперервної змінної.
    """
    plt.subplot(1, 2, subplot_position)
    plt.title(title)
    red_diamond = dict(markerfacecolor='r', marker='D')
    sns.boxplot(x=categorical,
                y=df[df[continuous] < max_continuous][continuous],
                data=df,
                flierprops=red_diamond,
                order=sorted(df[categorical].unique(), reverse=True),
                hue=hue_column, hue_order=sorted(df[hue_column].unique(), reverse=True))
    plt.ticklabel_format(style='plain', axis='y')
    plt.xticks(rotation=90)

def bi_boxplot(categorical, continuous, max_continuous1, max_continuous0, hue_column):
    """
    Створює паралельні блок-діаграми для двох груп, визначених у наборі даних, на основі
    категоріальної та неперервної змінної, виділяючи відмінності за допомогою відтінків.
    """
    plt.figure(figsize=(16, 10))

    # Графік для першо групи "Труднощі з платежами" (Payment Difficulties)
    draw_boxplot(df1, categorical, continuous, max_continuous1, 'Payment Difficulties', hue_column, 1)

    # Графік для другої групи "Вчасні оплати" (On-Time Payments)
    draw_boxplot(df0, categorical, continuous, max_continuous0, 'On-Time Payments', hue_column, 2)

    plt.tight_layout(pad=4)
    plt.show()
```
