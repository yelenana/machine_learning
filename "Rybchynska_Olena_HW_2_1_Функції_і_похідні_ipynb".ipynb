1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.





---



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f_x = 10 * x - 5
f_prime_x = sp.diff(f_x, x)
```

```python
print(f"Функція: f(x) = {f_x}")
print(f"Похідна функції: f'(x) = {f_prime_x}")
```

```python
x_values = np.linspace(0, 10, 100)
```

```python
f_numeric = sp.lambdify(x, f_x, 'numpy')
f_prime_numeric = sp.lambdify(x, f_prime_x, 'numpy')

y_values = f_numeric(x_values)
# Create an array of the derivative value with the same shape as x_values
y_prime_values = np.full_like(x_values, f_prime_numeric(x_values))

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label='$f(x) = 10x - 5$', color='blue')
plt.plot(x_values, y_prime_values, label="$f'(x) = 10$", linestyle='--', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("Кількість годин навчання (x)")
plt.ylabel("Оцінка / Значення похідної")
plt.grid(True)
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



---




```python
x = sp.symbols('x')
f_sin_x = sp.sin(x)
f_sin_prime_x = sp.diff(f_sin_x, x)
```

```python
print(f"Функція: f(x) = {f_sin_x}")
print(f"Похідна функції: f'(x) = {f_sin_prime_x}")
```

```python
x_values_sin = np.linspace(-2 * np.pi, 2 * np.pi, 400)
```

```python
# Перетворюємо символьні вирази на числові функції
f_sin_numeric = sympy.lambdify(x, f_sin_x, 'numpy')
f_sin_prime_numeric = sympy.lambdify(x, f_sin_prime_x, 'numpy')

y_sin_values = f_sin_numeric(x_values_sin)
y_sin_prime_values = f_sin_prime_numeric(x_values_sin)

plt.figure(figsize=(12, 6))
plt.plot(x_values_sin, y_sin_values, label='$f(x) = \sin(x)$', color='blue')
plt.plot(x_values_sin, y_sin_prime_values, label="$f'(x) = \cos(x)$", linestyle='--', color='red')
plt.title("Графік $f(x) = \sin(x)$ та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.axvline(0, color='gray', linewidth=0.8)
plt.axhline(0, color='gray', linewidth=0.8)
plt.show()
```



---



```python
f_ln_x = sp.log(x)
f_ln_prime_x = sp.diff(f_ln_x, x)
```

```python
print(f"Функція: f(x) = {f_ln_x}")
print(f"Похідна функції: f'(x) = {f_ln_prime_x}")
```

```python
x_values_ln = np.linspace(0.1, 5, 400)
```

```python
f_ln_numeric = sp.lambdify(x, f_ln_x, 'numpy')
f_ln_prime_numeric = sp.lambdify(x, f_ln_prime_x, 'numpy')

y_ln_values = f_ln_numeric(x_values_ln)
y_ln_prime_values = f_ln_prime_numeric(x_values_ln)

plt.figure(figsize=(12, 6))
plt.plot(x_values_ln, y_ln_values, label='$f(x) = \ln(x)$', color='green')
plt.plot(x_values_ln, y_ln_prime_values, label="$f'(x) = 1/x$", linestyle='--', color='purple')
plt.title("Графік $f(x) = \ln(x)$ та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.ylim(-3, 6)
plt.show()
```

3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.



---



```python
x = sp.symbols('x')
```

```python
f_cost_x = x**2 - 5*x + 1
```

```python
f_cost_prime_x = sp.diff(f_cost_x, x)
```

```python
print(f"Функція загальних витрат: f(x) = {f_cost_x}")
print(f"Похідна функції витрат (граничні витрати): f'(x) = {f_cost_prime_x}")
```

```python
f_cost_numeric = sp.lambdify(x, f_cost_x, 'numpy')
f_cost_prime_numeric = sp.lambdify(x, f_cost_prime_x, 'numpy')
```

```python
x_values = np.linspace(0, 100, 500)
```

```python
y_cost_values = f_cost_numeric(x_values)
y_cost_prime_values = f_cost_prime_numeric(x_values)
```

```python
plt.figure(figsize=(12, 7))
```

```python
plt.plot(x_values, y_cost_values, label='$f(x) = x^2 - 5x + 1$ (Загальні витрати)', color='blue')
plt.plot(x_values, y_cost_prime_values, label="$f'(x) = 2x - 5$ (Граничні витрати)", linestyle='--', color='red')

plt.title("Загальні та граничні витрати компанії")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати")
plt.grid(True)
plt.legend()
plt.show()
```
